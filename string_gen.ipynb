Generating random strings until a given string is generated

```python
import string
import random
import time
#possibleString=string.lowercase_ascii + string.digits + string.uppercase_ascill + ' ., !?;:'
possibleString= string.ascii_lowercase + string.digits + string.ascii_uppercase + ' ., !?;:'
t="colab"
attemptThis=''.join(random.choice(possibleString)
                            for i in range(len(t)))
attemptNext=""
completed=False
iterator=0
while completed==False:
  print(attemptThis)
  attemptNext=''
  completed=True
  for i in range(len(t)):
    if attemptThis[i]!=t[i]:
      completed=False
      attemptNext += random.choice(possibleString)
    else:
      attemptNext += t[i]

  iterator += 1
  attemptThis = attemptNext
  time.sleep(0.1)
print("Target match after:"+str(iterator) +"iterator")
```
